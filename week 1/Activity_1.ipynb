```python
import pandas as pd


df=pd.read_csv("Activity1.csv")
df.head()
```

```python
profitSeries = df['Profits']
stats = { 
        'Mean' : [profitSeries.mean()],
        'StDev' : [profitSeries.std()],
        'Variance': [profitSeries.var()],
        'Median' : [profitSeries.median()]
        }
pd.DataFrame(stats, index= ['Profits']).head()
```

```python
unprofitableDf = df[df['Profitable/Unprofitable'] == 2]
profitableDf = df[df['Profitable/Unprofitable'] == 1]
statsDf = pd.DataFrame({
        'Profitable/Unprofitable' : [1,2],
        'Mean' : [unprofitableDf['Profits'].mean(),profitableDf['Profits'].mean() ],
        'StDev' : [unprofitableDf['Profits'].std(),profitableDf['Profits'].std()], 
        'Variance': [unprofitableDf['Profits'].var(), profitableDf['Profits'].var()],
        'Median' : [unprofitableDf['Profits'].median(), profitableDf['Profits'].median()]
})
statsDf = statsDf.set_index('Profitable/Unprofitable')
statsDf.head()
```

```python
import pandas as pd
se = pd.Series([5,1,1,1,1,3,2,5,2,4])
pd.DataFrame({
    'Size' : [se.size],
    'Mean': [se.mean()],
    'Median': [se.median()],
    'Variance': [se.var()],
    'Standard_Deviation': [se.std()]
})
```

```python


numbers = []
def roll_d6():
  return random.randint(1, 6)

for _ in range(10):
    rollOutcome = roll_d6()
    numbers.append(rollOutcome)
print("roll outcomes: " + str(numbers))
se=pd.Series(numbers)
pd.DataFrame({
    'Size' : [se.size],
    'Mean': [se.mean()],
    'Median': [se.median()],
    'Variance': [se.var()],
    'Standard_Deviation': [se.std()]
})
```
